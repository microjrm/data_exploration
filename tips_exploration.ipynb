First things first, import the 'tips' dataset.  This was done using pandas.  I also have a tendency to want to see the firt few columns of an imported dataset, just to see if all headers are correct.

```python
import pandas as pd

#import csv and display data
tips = pd.read_csv('tips.csv')
display(tips.head())
```

I used seaborn for the first few visualizations for this dataset. It is all relatively simple and exploratory.  The main takeaway from my perspective was to see how these different types of visualizations were implemented and if any inferences could be made on the "tip life" of a restaurant worker. 

```python
import seaborn as sns
%matplotlib inline #this command forces plots to appear inline within the notebook.
```

I used the function regplot to show the linear relationship between total_bill and tip. It also shows the 95% confidence interval which I thought was handy.

```python
sns.regplot(x="total_bill", y="tip", data=tips)
```

I wanted to see how a strip plot would fair as we had one categorial variable

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

Box plot.

```python
sns.boxplot(x="day", y="total_bill", hue="time", data=tips)
```

Various bar plots for some exploration through the dataset.

```python
sns.barplot(x="sex", y="total_bill", hue="size", data=tips)
```

```python
sns.barplot(x="sex", y="total_bill", hue="time", data=tips)
```

```python
sns.barplot(x="sex", y="total_bill", hue="day", data=tips)
```

Hexbin plots. This visualization shows the counts of observations that fall within hexagonal bins. This plot works best with relatively large datasets.

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

Pairwise bivariate distribution

```python
sns.pairplot(tips)
```

Pairwise bivariate distribution, with the different colors representing lunch time and dinner time. 

```python
sns.pairplot(tips, hue="time")
```

Pairwise bivariate distribution, with the different colors representing male and female. 

```python
sns.pairplot(tips, hue = 'sex', markers = ["o", "s"])
```

# The next couple figures are from a combination of seaborn and matlabplot

regplot showing the relationship between tips and total bill with respect to the time of day and whether or not the server is a smoker.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,4))
fg = sns.FacetGrid(tips, row = "smoker", col = "time")
fg = fg.map(sns.regplot, "total_bill", "tip")
```

Repeated regplot, but this time with colors to differentiate between smoker/non-smoker

```python
plt.figure(figsize = (16,4))
fg = sns.FacetGrid(tips, row = "smoker", col = "time", hue = "smoker")
fg = fg.map(sns.regplot, "total_bill", "tip")
fg.add_legend()
```
